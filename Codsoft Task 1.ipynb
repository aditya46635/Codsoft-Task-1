```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x='Pclass',y='Survived',data=titanic_df)
plt.title('Survived vs Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Count values of Pclass
class_ = titanic_df['Pclass'].value_counts().reset_index()
class_.columns = ['Class', 'Total_Counts']

print(class_)

# Plot
plt.figure(figsize=(6, 3))
sns.barplot(x='Class', y='Total_Counts', data=class_, palette='Set3')
plt.title('Passenger Count by Class')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Filter only female passengers
female_passenger = titanic_df[titanic_df['Sex'] == 'female']  # note: usually lowercase in Titanic dataset
female_survived_counts = female_passenger['Survived'].value_counts()

# Extract counts
num_female_survived = female_survived_counts.get(1, 0)   # survived = 1
num_female_died = female_survived_counts.get(0, 0)       # not survived = 0

print("Female Survived:", num_female_survived)
print("Female Not Survived:", num_female_died)

plt.figure(figsize=(6, 3))
sns.barplot(x=['Survived', 'Not Survived'],y=[num_female_survived, num_female_died],palette='Set2')
plt.title('Female Passenger Survived vs Not Survived')
plt.xlabel('Status')
plt.ylabel('Number of Passengers')
plt.show()
```
